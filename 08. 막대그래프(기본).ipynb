# 8.막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [189, 187, 184] #키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [189, 187, 184] #키

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [189, 187, 184] #키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [189, 187, 184] #키

plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) #X 축의 이름 데이터 각도를 45도로 회전
plt.yticks(rotation=45) #Y 축의 이름 데이터 각도를 45도로 회전
```

```python
labels = ['강백호', '서태웅', '정대만'] #이름
values = [189, 187, 184] #키
ticks=['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.ticks(labels, ticks)
```
